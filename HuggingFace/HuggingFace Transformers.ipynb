### 1.Load and Prepare dataset

```python
from datasets import load_dataset
dataset = load_dataset('imdb')
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset['train'][0]
```

```python
dataset['test'][0]
```

```python
dataset['unsupervised'][0]
```

### 2. Data Preprocessing

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
```

```python
# Tokenize the dataset

def tokenize_ds(text):
  return tokenizer(text['text'],padding = 'max_length',truncation = True)
```

```python
tokenized_ds = dataset.map(tokenize_ds,batched = True)
```

```python
tokenized_ds
```

### 3. Training the Dataset

```python
from transformers import TrainingArguments
```

```python
training_args = TrainingArguments(
    output_dir = './results',
    eval_strategy = 'epoch',
    learning_rate = 2e-5,
    per_device_train_batch_size = 16,
    per_device_eval_batch_size = 16,
    num_train_epochs = 1,
    weight_decay = 0.01,

)
```

```python
training_args
```

### 4.  Initializing and Training the model

```python
import torch
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
device
```

```python
from transformers import AutoModelForSequenceClassification , Trainer
model = AutoModelForSequenceClassification.from_pretrained('bert-base-cased',num_labels = 2).to(device)
```

```python
model_trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset = tokenized_ds['train'],
    eval_dataset = tokenized_ds['test']
)
```

```python
model_trainer.train()
```

### 5. Evaluate and save the Model

```python
outputs =  model_trainer.evaluate()
```

```python
model.save_pretrained('./fine_tuned_model')
tokenizer.save_pretrained('./fine_tuned_tokenizer')
```

## Arvix Project
`Arvix` -- arXiv," which is a repository for research papers in fields like physics, mathematics, computer science, and more. arXiv allows researchers to share their preprints (drafts of their research papers) before they are formally peer-reviewed and published in journals.

```python
!pip install arxiv
```

```python
import arxiv
import pandas as pd
```

```python
query = 'ai OR artificial intelligence OR machine learning'
search = arxiv.Search(query=query, max_results=10, sort_by=arxiv.SortCriterion.SubmittedDate)
```

```python
papers = []
for result in search.results():
    papers.append({
      'published': result.published,
        'title': result.title,
        'abstract': result.summary,
        'categories': result.categories
    })

# Convert to DataFrame
df = pd.DataFrame(papers)

pd.set_option('display.max_colwidth', None)
df.head(10)
```

```python
abstract = df['abstract'][0]
abstract
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
summarization_result = summarizer(abstract)
```

```python
summarization_result[0]['summary_text']
```
